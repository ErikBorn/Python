```python
import matplotlib.pyplot as plt
import numpy as np

# Define x range
x = np.linspace(0, 45, 200)

# Define slopes (m) and intercepts (b) for four lines
slopes = [979,792,1927,881]      # <-- replace with your values
intercepts = [53181,60099,56267,57757]      # <-- replace with your values
labels = ["BA", "MA", "PhD", "Total"]  # <-- replace with your labels


# Plot each line
for m, b, label in zip(slopes, intercepts, labels):
    y = m * x + b
    plt.plot(x, y, label=label)

# Add labels, legend, and grid
plt.xlabel("Years of Experience")
plt.ylabel("Salary ($)")
plt.title("Salary function comparisons")
plt.legend()
plt.grid(True)

plt.show()
```
