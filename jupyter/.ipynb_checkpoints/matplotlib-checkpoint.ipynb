```python
xx = np.linspace(-2*np.pi,2*np.pi,1000)
fig, ax = plt.subplots(dpi = 100)
ax.plot(xx,np.cos(xx),'b.')
ax.plot(xx,np.sin(xx),'g.')
ax.legend(["cos(x)",'sin(x)'])
ax.set_xlabel = "angle (rad)"
ax.set_ylabel = "trig value"
ax.plot([-2*np.pi,2*np.pi],[0,0],'k')
```

```python
fig, Axes = plt.subplots(1,2,figsize = (10,5))
Axes[0].plot(x,x**2,'b.',x,x**3,'b.',lw = 2)
Axes[0].grid(True)
Axes[1].plot(x,x**2,'b.',x,x**3,'b.',lw = 2)
Axes[1].grid(True)
Axes[1].set_ylim([0,60])
Axes[1].set_xlim([2,5])

```

```python
n = np.arange(0,6,1)
fig, axes = plt.subplots(2,2,figsize = (16,5))
axes[0][0].set_title('scatter')
axes[0][0].scatter(x,x+.25*np.random.randn(len(x)))

axes[0][1].set_title('step')
axes[0][1].step(n,n**2,lw=2)

axes[1][0].set_title('bar')
axes[1][0].bar(n,n**2,align='center')

axes[1][1].set_title('fill between ')
axes[1][1].fill_between(x,x**2.5,x**2.7,color='blue',alpha = 0.5)
```

```python
#Radar chart
fig = plt.figure(figsize = (5,5))
ax = fig.add_axes([0.,0.,.6,.6], polar = True)
t = np.linspace(0,2*np.pi,100)
ax.plot(t,np.cos(t),'b.')
```

```python
#hists
n = np.random.randn(100000)
fig, axes = plt.subplots(1,2,figsize=(12,4))
axes[0].set_title('default')
axes[0].hist(n,bins = 50)

axes[1].set_title('other')
axes[1].hist(n,cumulative = True, bins = 50)
```

```python
#draw contour
import matplotlib.cm as cm
delta = 0.25
xxx = np.arange(-3,3,delta)
yyy = np.arange(-2,2,delta)
X, Y = np.meshgrid(xxx,yyy)
Z1 = np.exp(-X**2 - Y**2)
Z2 = np.exp(-(X-3)**2-(Y-3)**2)
Z = (Z1-Z2)*2
fig, ax = plt.subplots()
CS = ax.contour(X, Y, Z)
ax.clabel(CS,inline = 1, fontsize =10)
ax.set_title('Contour')
```

```python
#Drawing 3d
from mpl_toolkits.mplot3d.axes3d import Axes3D as a3d

fig = plt.figure(figsize=(14,6))

ax = fig.add_subplot(1,2,1,projection='3d')
ax.plot_surface(X,Y,Z,rstride = 1, cstride = 1, linewidth = 0)
```
